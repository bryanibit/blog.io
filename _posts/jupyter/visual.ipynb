# Test accuracy for xsens and novatel

```python
import matplotlib.pyplot as plt
# import math
import numpy as np
nova_ll = []
nova_angle=[]
xsens = []
class origin:
    def __init__(self):
        self.time=1503483280
        self.originx=623577.0
        self.originy=4325012.0
with open('novatel_ll.txt','r') as novatel_ll:
    for line in novatel_ll.readlines():
        nova_ll.append([float(i) for i in line.strip().split(',')])
        
with open('novatel_angle.txt','r') as novatel_angle:
    for line in novatel_angle.readlines():
        nova_angle.append([float(i) for i in line.strip().split(',')])
        
with open('xsens.txt','r') as xsens_fl:
    for line in xsens_fl.readlines():
        xsens.append([float(i) for i in line.strip().split(',')])
ori=origin()
plt.figure('novatel_xsens_ll',figsize=(12,7))
plt.plot(np.array(nova_ll)[:,1]-ori.originx,np.array(nova_ll)[:,2]-ori.originy,'r-')
plt.plot(np.array(xsens)[:,1]-ori.originx,np.array(xsens)[:,2]-ori.originy,'b-')

# plt.figure('novatel_ll & xsens_ll')
plt.show()
```

```python
plt.figure('xsens_novatel_yaw',figsize=(12,7))
plt.plot(np.array(nova_angle)[:,0]-ori.time,np.array(nova_angle)[:,1]-np.floor(np.array(nova_angle)[:,1]/180)*360,'r-')
plt.plot(np.array(xsens)[:,0]-ori.time,np.array(xsens)[:,3]*(-1),'b-')
plt.show()
```

```python
plt.figure('xsens_novatel_pitch',figsize=(12,7))
plt.plot(np.array(nova_angle)[:,0]-ori.time,np.array(nova_angle)[:,2],'r-')
plt.plot(np.array(xsens)[:,0]-ori.time,np.array(xsens)[:,4]*(-1),'b-')
plt.show()
```

```python
plt.figure('xsens_novatel_roll',figsize=(12,7))
plt.plot(np.array(nova_angle)[:,0]-ori.time,np.array(nova_angle)[:,3],'r-')
plt.plot(np.array(xsens)[:,0]-ori.time,np.array(xsens)[:,5]*(-1),'b-')
plt.show()
```
